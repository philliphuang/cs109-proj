# Notebook 10: Advanced Models Part 2 (Log)
Now that we have explored Multiple Linear Regression as well as Lasso, let's explore different follow-up methods to address the issues we noticed before.

```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
import pickle
from collections import Counter
import sklearn as sk
from sklearn.cross_validation import LeaveOneOut
import StringIO
from sklearn.preprocessing import Imputer
from sklearn import linear_model
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor as rf
%matplotlib inline
from sklearn.grid_search import GridSearchCV
```

```python
# Load cleaned NCD data from pickled files
out = open('data/clean/deaths_100k.p', 'r')
deaths_100k = pickle.load(out)
out.close()
out = open('data/clean/risk.p', 'r')
risk_of_death = pickle.load(out)
out.close()
out = open('data/clean/crops.p', 'r')
crops = pickle.load(out)
out.close()
out = open('data/clean/meat.p', 'r')
meat = pickle.load(out)
out.close()
out = open('data/clean/var_desc_livestock.p', 'r')
livestock_desc = pickle.load(out)
out.close()
out = open('data/clean/var_desc_crops.p', 'r')
crops_desc = pickle.load(out)
out.close()

# load cleaned food data with redundant variables removed
out = open('data/final/food_1970_2000_cleaned.p', 'r')
food_1970_2000_cleaned = pickle.load(out)

# Get response variables
deaths_100k_all_2000 = deaths_100k['all'][2000]
deaths_100k_cancer_2000 = deaths_100k['cancer'][2000]
deaths_100k_cardio_2000 = deaths_100k['cardio'][2000]
deaths_100k_diabetes_2000 = deaths_100k['diabetes'][2000]
deaths_100k_resp_2000 = deaths_100k['resp'][2000]
deaths_100k_all_2012 = deaths_100k['all'][2012]
deaths_100k_cancer_2012 = deaths_100k['cancer'][2012]
deaths_100k_cardio_2012 = deaths_100k['cardio'][2012]
deaths_100k_diabetes_2012 = deaths_100k['diabetes'][2012]
deaths_100k_resp_2012 = deaths_100k['resp'][2012]

risk_of_death_2000 = risk_of_death[2000]
risk_of_death_2012 = risk_of_death[2012]

time_period = range(1970, 2000)

# Calculate the mean for each crop/meat over the period 1970-2000
food_1970_2000 = pd.DataFrame(index=risk_of_death.index)

for crop in crops.iterkeys():
    food_1970_2000[crop] = crops[crop][time_period].mean(axis=1)
    
for m in meat.iterkeys():
    food_1970_2000[m] = meat[m][time_period].mean(axis=1)

# list of countries to drop due to being less than 50% full
countries_to_drop = []

for index, row in food_1970_2000.iterrows():
    if row.isnull().sum() > len(row) / 2:
        countries_to_drop.append(index)
        
# drop countries from response variables
deaths_100k_all_2000 = deaths_100k_all_2000.drop(countries_to_drop)
deaths_100k_cancer_2000 = deaths_100k_cancer_2000.drop(countries_to_drop)
deaths_100k_cardio_2000 = deaths_100k_cardio_2000.drop(countries_to_drop)
deaths_100k_diabetes_2000 = deaths_100k_diabetes_2000.drop(countries_to_drop)
deaths_100k_resp_2000 = deaths_100k_resp_2000.drop(countries_to_drop)

deaths_100k_all_2012 = deaths_100k_all_2012.drop(countries_to_drop)
deaths_100k_cancer_2012 = deaths_100k_cancer_2012.drop(countries_to_drop)
deaths_100k_cardio_2012 = deaths_100k_cardio_2012.drop(countries_to_drop)
deaths_100k_diabetes_2012 = deaths_100k_diabetes_2012.drop(countries_to_drop)
deaths_100k_resp_2012 = deaths_100k_resp_2012.drop(countries_to_drop)

risk_of_death_2000 = risk_of_death_2000.drop(countries_to_drop)
risk_of_death_2012 = risk_of_death_2012.drop(countries_to_drop)
```

```python
# Plots bar graph of coefficient estimates for regression model and given significance level 
def plot_sig_coeffs(fit, sig_level):
    # Filter for predictors with p-values less than provided cutoff
    significant = fit.pvalues[fit.pvalues <= sig_level]
    # Plot coefficient estimates for predictors with p < 0.1 
    fit.params[significant.index].sort_values().plot(kind="bar", 
        title="Coefficient Estimates for Predictors with p < 0.1", 
        )
```

# Log Transform
Now let's try performing a log transformation on the response variables and see if that addresses the absence of normality we noticed in our earlier models. Note that we only perform this transformation for the age-standardized mortality rates because risk of death is calculated as a percentage already, so there would be no need to transform that data. First, however, we should check if log transformations would make sense for each dataset.

```python
deaths_datasets = [deaths_100k_all_2000, deaths_100k_cancer_2000, deaths_100k_cardio_2000, 
                   deaths_100k_diabetes_2000, deaths_100k_resp_2000]
disease_types = ['All', 'Cancer', 'Cardiovascular', 'Diabetes', 'Respiratory']
for i in range(len(deaths_datasets)):    
    plt.hist(deaths_datasets[i], bins = 'auto')
    plt.title(disease_types[i])
    plt.tight_layout()
    plt.show()
```

Since all of the datasets with age-standardized mortality rates are right-skewed, log transformations should be appropriate for all of them.

```python
# applying log transformations to the relevant datasets
deaths_100k_all_2000_log = np.log(deaths_100k_all_2000)
deaths_100k_cancer_2000_log = np.log(deaths_100k_cancer_2000)
deaths_100k_cardio_2000_log = np.log(deaths_100k_cardio_2000)
deaths_100k_diabetes_2000_log = np.log(deaths_100k_diabetes_2000)
deaths_100k_resp_2000_log = np.log(deaths_100k_resp_2000)
```

```python
# checking the historgrams once again
deaths_datasets_log = [deaths_100k_all_2000_log, deaths_100k_cancer_2000_log, deaths_100k_cardio_2000_log, 
                   deaths_100k_diabetes_2000_log, deaths_100k_resp_2000_log]
for i in range(len(deaths_datasets_log)):
    plt.hist(deaths_datasets_log[i], bins = 'auto')
    plt.title(disease_types[i])
    plt.tight_layout()
    plt.show()
```

Most of the histograms now appear more Normally distributed, although the transformed responses for all non-communicable diseases and cardiovascular disease look potentially bimodal which could be an issue. Still, we can check the results and diagnostics of different models under the new transformations to see if they make more sense.

# Multiple Linear and Lasso Regression under Log Transformations
Now we will fit both Multiple Linear Regression and Lasso Regression models for each of the categories of response variables. Once again we use GridSearchCV to select the best hyperparameters for the Lasso models for each response variable, using Leave One Out cross validation to account for overfitting. While K-Fold cross validation is normally the preferred method for its computational efficiency, for our data it could vary significantly depending on the fold choice.

## Deaths per 100k from all non-communicable diseases

```python
lasso = linear_model.Lasso(alpha = 1.0)
LOO = LeaveOneOut(n = deaths_100k_all_2000_log.shape[0])
lasso.fit(food_1970_2000_cleaned, deaths_100k_all_2000_log)
y_deaths_100k_all= deaths_100k_all_2000_log.sort_index().loc[food_1970_2000_cleaned.index].values
lasso.score(food_1970_2000_cleaned, y_deaths_100k_all)
params = {
    'alpha': [0.3,0.6,1.0,1.3,1.6,2.0,2.3,2.6,3.0,4.0,5.0]
}

CV_model = GridSearchCV(lasso, param_grid=params, cv=LOO)
CV_model.fit(food_1970_2000_cleaned, y_deaths_100k_all)
CV_model.best_params_
```

With the best parameters identified, we now fit both the Multiple Linear Regression and Lasso models accordingly.

```python
deaths_all_model = sm.OLS(y_deaths_100k_all, food_1970_2000_cleaned)
deaths_all_results = deaths_all_model.fit()
deaths_all_results_lasso = deaths_all_model.fit_regularized(alpha = 0.3)
deaths_all_results.summary()
```

Even after the logarithmic transformation on the age-standardized mortality rate from all non-communicable diseases, we still reach the same conclusion from the Omnibus and Jarque-Bera tests that the residuals are not normally distributed. Moreover, the p values were very similar for the model of the transformed mortality rate compared to the regular model, indicating that the log transformation did little to address the issue.

```python
plot_sig_coeffs(deaths_all_results, 0.1)
```

Now let's check whether the Lasso regression model is improved by the transformation.

```python
deaths_all_results_lasso.summary()
```

Under lasso regression, the log transformation seemed to exacerbate the issue of non-normality of the residuals, so we should try a different approach.

```python
plot_sig_coeffs(deaths_all_results_lasso, 0.1)
```

Now we use the same approach for the datasets considering the age-standardized mortality rates from each individual non-communicable disease as well.

## Deaths per 100k from Cancer

```python
lasso = linear_model.Lasso(alpha = 1.0)
LOO = LeaveOneOut(n = deaths_100k_cancer_2000_log.shape[0])
lasso.fit(food_1970_2000_cleaned, deaths_100k_cancer_2000_log)
y_deaths_100k_cancer = deaths_100k_cancer_2000_log.sort_index().loc[food_1970_2000_cleaned.index].values
lasso.score(food_1970_2000_cleaned, y_deaths_100k_cancer)
params = {
    'alpha': [0.3,0.6,1.0,1.3,1.6,2.0,2.3,2.6,3.0,4.0,5.0],
}

CV_model = GridSearchCV(lasso, param_grid=params, cv=LOO)
CV_model.fit(food_1970_2000_cleaned, y_deaths_100k_cancer)
CV_model.best_params_
```

```python
deaths_cancer_model = sm.OLS(y_deaths_100k_cancer, food_1970_2000_cleaned)
deaths_cancer_results = deaths_cancer_model.fit()
deaths_cancer_results_lasso = deaths_cancer_model.fit_regularized(alpha = 0.3)
deaths_cancer_results.summary()
```

Likewise, when considering the logarithmically transformed age-standardized mortality rate from cancer, the Omnibus and Jarque-Bera Tests reject the null hypothesis of normality in the residuals. However, the test statistics became much less extreme and the p values approached the significance threshold of 0.05, so the log transformation appears to have addressed the normality issue somewhat.

```python
plot_sig_coeffs(deaths_cancer_results, 0.1)
```

```python
deaths_cancer_results_lasso.summary()
```

Similarly, while the Omnibus and Jarque-Bera tests still claim enough evidence to reject the null hypothesis of normality in the residuals in the lasso model, their respective test statistics are less extreme, indicating that some of the problem was addressed. 

```python
plot_sig_coeffs(deaths_cancer_results_lasso, 0.1)
```
